### (i)Handling Missing Values
The **DateofTermination** column contains missing values for employees who are currently employed and have not been terminated. Since these missing values have a valid business meaning, they were retained as null values (NaT) rather than being imputed. This preserves the integrity of the date field while accurately representing active employees.
I am going to create a new column as **Employee_Status** where **Active** indicates **Current** employees and **Inactive** indicates **Ex-Employees**.

```python
#Creating a new column for Active and Inactive Employees
df['Employee_Status'] = df['DateofTermination'].isna().map({
    True: 'Active',
    False: 'Terminated'
})
df.head()
```

**Info:-*** The Employee Status Column represents the current status of the employees on the organization, based on the termination date column, if an employee contains the value on the terminatin date field it indicates that the employee has been terminated from the organization, and if not the employee is still part of the organization.

#### (iii) Bivariate Analysis
It examines the relationship between two variables to identify patterns,associations,and differences.It helps understand how one feature may influence or vary with another.

```python
# function to summarize  HR data
def Attrition_Stats(column):
    hr = df.groupby(column, observed=True).agg(
        Attrition_Count=("Employee_Status",
                         lambda x: (x == "Terminated").sum())
    )
    return hr

Attrition_Stats("Department")
```

***Info:-*** **Groups the data** by a chosen column by using **groupby** function and calculates the attrition count  for each group.

```python
#Department wise Attrition status
Attrition_Stats("Department")
```

```python
#Gender wise Attrition status
Attrition_Stats("Sex")
```

```python
#Recruitment source wise Attrition status
Attrition_Stats("RecruitmentSource")
```

### (i) Employee Distribution by Department
• ***Visualization Type:-*** Donut Chart  

• ***Purpose:-*** Analyzes the number of employees working in each department to understand  workforce allocation.  

• ***Insight:-*** The employee distribution varies across departments, indicating that certain departments require a larger workforce to support core business operations. Departments with higher employee counts represent key functional areas, while departments with fewer employees may focus on specialized responsibilities. This analysis helps identify workforce concentration and supports resource planning and staffing decisions.

```python
plt.figure(figsize=(8,8))

plt.pie(
    df['Department'].value_counts(),
    labels=df['Department'].value_counts().index,
    autopct='%1.1f%%'
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Employee Distribution by Department")
plt.show()
```
